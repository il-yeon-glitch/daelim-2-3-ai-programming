```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "study_hours": [1, 2, 3, 3, 4, 4, 5, 6, 7, 8],
    "attendance": [55, 55, 60, 80, 60, 75, 75, 80, 90, 95],
    "pass_fail": [0, 0, 0, 1, 0, 1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# 입력값과 정답 분리
X = df[["study_hours", "attendance"]]
y = df["pass_fail"]

model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

print("데이터 준비 완료")
print(f"학생 수: {len(df)}명") # 학생 수는: %%명
```

# 모델 정확도

- 학습에 필요한 데이터로 정확도 확인

```python
accuracy = model.score(X, y)
print("모델 정확도:", accuracy) # 0 ~ 1 표시됨 -> 0 ~ 100%
print(f"모델 정확도 퍼센트: {accuracy*100}%")
```

```python
# 모델이 어떤 예측을 했는지 확인
predictions = model.predict(X)

# Columns - 공부시간, 출석률, 실제, 예측, 결과
# zip() -> zip([1, 2, 3], [4, 5, 6]) -> (1,4), (2,5), (3,6) // 각각 배열의 크기가 같아야함
# 실제값과 예측값을 한쌍으로 묶어서
result = pd.DataFrame({
    "공부시간": X["study_hours"].values,
    "출석률": X["attendance"].values,
    "실제": y.values,
    "예측": predictions,
    "결과": ["0" if a == p else "X" for a, p in zip(y, predictions)]
    # zip(실제값, 예측값), 배열 처리, 람다 표현식(일반적인 if문을 한 줄로 표현)
})

result
```

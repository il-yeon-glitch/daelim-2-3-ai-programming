```python
def clacilate_avgrage(numbers):
    total = 0
    for n in numbers:
        total += n
    return total / len(numbers)

my_socres = [80, 90, 70, 100]
avg = clacilate_avgrage(my_socres)
print("평균:", avg)
```

복습 문제 - 종합

3명의 학생 이름과 점수가 있고 점수에 따라 등급을 매기는 함수를 생성하는 로드를 출력합니다.

A 학생은 A 등급입니다
B 학생은 B 등급입니다
C 학생은 C 등급입니다

```python
# 문제

students = ["A", "B", "C"]
students_socres = [95, 58, 73]

def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

for name, score in zip(students, students_socres):
    grade = get_grade(score)
    print(f"{name} 학생은 {grade} 등급입니다")
```

# 머신러닝

> 정의: 사람이 규칙을 직접 생성하는 대신,   
> 컴퓨터가 데이터를 보고 컴퓨터가 스스로 규칙을 찾아내는 방법

일반 프로그래밍 vs 머신러닝

| 구분 | 일반 프로그래밍 | 머신러닝 |
| --- | --- | --- |
| 규칙을 누가 만드나 | 사람이 직접 만든다 | 컴퓨터가 데이터를 찾는다 |
| 필요한 것 | 규칙 (if문 등) | 데이터 + 정답 |
| 예시 | if 점수 >= 60: 합격 | 과거 데이터를 참조, 기준을 찾게함 |
| 장점 | 규칙이 적거나, 명확할 때 좋음 | 규칙이 많거나, 복잡할 때 좋음 |

# 모델 학습과 예측 실습

## 모델이란?

> 정의: 데이터를 보고 규칙을 학습한 뒤,   
> 새로운 데이터가 들어오면 결과를 예측하는 프로그램 

쉬운 비유:
- 모델 = 학생
- 학생(모델)에게 문제와 답, 개념(데이터)을 주고 공부를 시킵니다.
- 공부가 끝나면, 새로운 문제(데이터)를 풀게 합니다.

## 학습(fit)이란?

> 정의: 모델이 입력값(x)과 정답(y)을 보고   
> 둘 사이의 패턴이나 규칙을 찾는 과정

- `model.fit(x, y)`을 사용하면 학습이 시작됩니다.
- x(입력값)와 y(정답)을 함께 넣어줘야 합니다.
- 모델은 학습이 끝나고 규칙을 기억하고 있게 됩니다.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "study_hours": [1, 2, 3, 3, 4, 4, 5, 6, 7, 8],
    "attendance": [55, 55, 60, 80, 60, 75, 75, 80, 90, 95],
    "pass_fail": [0, 0, 0, 1, 0, 1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# 입력값과 정답 분리
x = df[["study_hours", "attendance"]]
y = df["pass_fail"]

print("데이터 준비 완료")
print(f"학생 수: {len(df)}명") # 학생 수는: %%명
```

```python
# 모델 만들기

model = DecisionTreeClassifier(random_state=42) 
model.fit(x, y)
print("모델 학습 완료")
```

```python
# 새 학생 1명을 예측해보기
# 주당 4시간 공부, 출석률 80%

new_student = pd.DataFrame([[4, 80]], columns=["study_hours", "attendance"])
prediction = model.predict(new_student)
print("공부 4시간, 출석률 80%인 학생:", prediction)

if prediction[0] == 1:
    print("모델 판단: 합격")
else:
    print("모델 판단: 불합격")
```

```python
# 여러 학생 예측하기
# 4명의 학생 데이터 삽입
new_students = pd.DataFrame([
    [4, 80],
    [3, 65],
    [5, 80],
    [7, 95],
    ], columns=["study_hours", "attendance"])

results = model.predict(new_students)

for i in range(len(new_students)):
    study = new_students.iloc[i, 0]
    attendance = new_students.iloc[i, 1]
    
    if results[i] == 1:
        status = "합격"
    else:
        status = "불합격"
        
    print(f'공부 {study}시간, 출석 {attendance}% 인 학생은 "{status}" 입니다.')

# 문제:
# 각 학생의 로그를 출력해주세요.

# 예시:
# 공부 ?시간, 출석 ?% 인 학생은 "합격" 입니다.
# 공부 ?시간, 출석 ?% 인 학생은 "불합격" 입니다.
```

```python
scores = [42, 60, 77, 89, 20]

for score in scores:
    if score >= 60:
        print(f'{score}점은 합격입니다')
    else:
        print(f'{score}점은 불합격입니다')
```

```python
scores = [65, 82, 91, 100, 58, 77]

count = 0

for score in scores:
    if score >= 70:
        print(f'{score}점은 합격입니다')
        count+=1
    else:
        print(f'{score}점은 불합격입니다')

print(f'합격자는 총 {count}입니다')
```
